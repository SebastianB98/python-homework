```python
import pandas as pd
from pathlib import Path
Path_file = Path('../Python_Homework_Resources/Homework_02-Python_Instructions_PyBank_Resources_budget_data.csv')
budget_data_csv = pd.read_csv(Path_file)
total_months = len(budget_data_csv['Date'])
print('Total Months: ' + str(total_months))
total_profit_and_losses = budget_data_csv['Profit/Losses'].sum()
print('Total: ' + "$" + str(total_profit_and_losses)) 
AmtChange = budget_data_csv["Profit/Losses"].diff()
average_change = budget_data_csv['Profit/Losses'].pct_change().sum()
print(f'Average Change: {average_change}')
budget_data_csv["Amount Changed"] = AmtChange
greatest_increase = budget_data_csv['Amount Changed'].max()
print(f'Greatest Increase in Profits: {greatest_increase}')
greatest_decrease = budget_data_csv['Amount Changed'].min()
print(f'Greatest Decrease in Profits: {greatest_decrease}')
with open("PyBank.txt","w") as text_file:
    text = 'Total Months: ' + str(total_months) + '\n' 'Total: ' + "$" + str(total_profit_and_losses) + '\n' f'Average Change: {average_change}' + '\n' f'Greatest Increase in Profits: {greatest_increase}' + '\n' f'Greatest Decrease in Profits: {greatest_decrease}'
    text_file.write(str(text))
    text_file.close()

```
